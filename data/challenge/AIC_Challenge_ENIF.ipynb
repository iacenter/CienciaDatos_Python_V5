```python
#gift function without comments

def preguntas_opcion_multiple(df, lista_columnas):
  df_result = pd.DataFrame(columns = ["value", "count"])
  for p in lista_columnas:
    a = df.groupby(p, as_index=False).size()
    a = a[a[p] != "No se declaró como opción afirmativa"]
    a = a.rename(columns={p: "value", "size": "count"})
    df_result = pd.concat([df_result, a])
  return df_result
```
